```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
#Set type distributions here
std_1a = 0.50
mean_1a = -19.25
std_1b = 1.12
mean_1b = -17.45
std_1c = 1.18
mean_1c = -17.66
std_2b = 0.92
mean_2b = -16.99
std_2l = 0.86
mean_2l = -17.98
std_2n = 1.36
mean_2n = -18.53
std_2p = 0.98
mean_2p = -18.53
```

```python
#Defines the distribution, for plotting
def f(M,std,mean):
    f=(1./(2.*np.pi*std**2.)**(1./2.))*np.exp(-(M-mean)**2./(2.*std**2))
    return f
```

```python
M=np.linspace(-14,-23,1000)
```

```python
#print(M)
```

```python
y_1a=f(M,std_1a,mean_1a)
y_1b=f(M,std_1b,mean_1b)
y_1c=f(M,std_1c,mean_1c)
y_2b=f(M,std_2b,mean_2b)
y_2l=f(M,std_2l,mean_2l)
y_2n=f(M,std_2n,mean_2n)
y_2p=f(M,std_2p,mean_2p)
```

```python
plt.plot(M,y_1a,M,y_1b,M,y_1c,M,y_2b,M,y_2l,M,y_2n,M,y_2p)
```

```python
from scipy.integrate import quad
```

```python
#For calculating the luminosity distance, formula in notes
def integrand(x):
    return 1.0/(0.308*(1+x)**(3.0)+0.692*(1+x)**(6.0))**(0.5)
```

```python
#Set redshift here!
z=1.0
c=3e5
H_0=67.8
```

```python
I = quad(integrand, 0, z)
```

```python
I, I[0], I[1]
```

```python
d_L = ((1.0+z)*c)/H_0*I[0]
```
