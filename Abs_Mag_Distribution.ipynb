```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
std = 0.50
mean = -19.25
```

```python
def f(M,std,mean):
    f=(1./(2.*np.pi*std**2.)**(1./2.))*np.exp(-(M-mean)**2./(2.*std**2))
    return f
```

```python
M=np.linspace(-15,-22,600)
```

```python
#print(M)
```

```python
y=f(M,std,mean)
```

```python
plt.plot(M,y)
```

```python
from scipy.integrate import quad
```

```python
def integrand(x):
    return 1.0/(0.3*(1+x)**(3.0)+0.7*(1+x)**(6.0))**(0.5)
```

```python
z=0.0005
c=3e5
```

```python
I = quad(integrand, 0, z)
```

```python
I
```

```python
H_0=67.7
```

```python
d_L = ((1.0+z)*c)/H_0*I[0]
```

```python
d_L
```

```python
m_mean = mean + 5.0*np.log10(d_L)+25.0
```
