```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
%matplotlib inline
```

```python
std = 0.50
mean = -19.25
std2 = 1.36
mean2 = -18.53
```

```python
def f(M,std,mean):
    f=(1./(2.*np.pi*std**2.)**(1./2.))*np.exp(-(M-mean)**2./(2.*std**2))
    return f
```

```python
M=np.linspace(-15,-22,600)
```

```python
y=f(M,std,mean)
y2=f(M,std2,mean2)
```

```python
plt.plot(M,y)
```

```python
plt.plot(M,y2,M,y)
```

```python
s = np.random.normal(mean, std, 600)
s2= np.random.normal(mean2,std2,600)
```

```python
count, bins, ignored = plt.hist(s, 30, normed=True)
count, bins, ignored = plt.hist(s2,30, normed=True)
plt.plot(M,y,M,y2)
```

```python
sn = np.random.normal(mean,std,1000)
```

```python
count, bins, ignored = plt.hist(sn,50,normed=True)
plt.plot(bins, 1/(std*np.sqrt(2*np.pi))*
        np.exp( - (bins-mean)**2/(2*std**2)),
        linewidth=2, color='r')
plt.show()
```
